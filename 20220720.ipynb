> ### Gradient Descent Algorithm

시작하는 포인트에 따라서 local optimum을 달성하느냐 혹은 global optimum을 달성하느냐가 달려있음

-> 개선이 필요함

-> Momentum을 활용하자
- 
- 
- 현재의 momentum을 과거의 momentum을 축적
- 과거의 값을 반영한 gradient
> ##### 설령 local minimum이나 saddle point에서 gradient가 0이 되는 지점이 발생해도 과거에 이어오던 momentum값을 반영해주기 때문에 계속 학습이 가능한 것!

> ### AdaGrad

수렴의 속도를 조절하는 것

단점??
- gradient값이 누적됨에 따라 learning rate이 작아지게 됨
- 즉, 학습이 그 지점에서 일어나지 않게 됨


대책??
- RMSProp algorithm
  - 종전처럼 gradient값이 누적됨에 따라 theta값이 줄어드는 것이 아니라 완충된 속도로 학습이 이루어진다고 생각하면 됨
